```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./iris.csv")
dataset.columns = [
    "Id",
    "sepal length (cm)", 
    "sepal width (cm)", 
    "petal length (cm)", 
    "petal width (cm)", 
    "species"]
```

```python
dataset.head()
```

```python
# Transform the species to numeric values:
```

```python
mappings = {
   "Iris-setosa": 0,
   "Iris-versicolor": 1,
   "Iris-virginica": 2
}
```

```python
dataset["species"] = dataset["species"].apply(lambda x: mappings[x])
```

```python
dataset.head()
```

```python
# The below two lines will separate the species (true label) and rest of the columns
X = dataset.drop("species",axis=1).values
y = dataset["species"].values

print(X)
print(y)
```

```python
# This shuffles the contents.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```
